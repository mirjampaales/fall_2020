# Scatterplots

Lets try to make some scatterplots to see the shapes better. (and see that they are linear-ish)

# Math exam and GPA

## Linear

```python
#sns.scatterplot(x="math_exam", y="grade_no", data=all_data);
sns.lmplot(x="math_exam", y="fall GPA", data=final_data, hue = "GENDERMALE",  x_jitter=.15);
```

## Second order polynomial

```python
sns.lmplot(x="math_exam", y="fall GPA", data=final_data, order = 2, hue = 'GENDERMALE',  x_jitter=.15);
```

##  Log-linear

```python
sns.lmplot(x="math_exam", y="fall GPA", data=final_data, hue = 'GENDERMALE', logx=True,  x_jitter=.15);
```
